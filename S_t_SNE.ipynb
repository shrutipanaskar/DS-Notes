```python
#from pandas import read_csv
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
#load data
df = pd.read_csv('TSNE_data.csv')
```

```python
df.head() #M Malignant - Cancer tumor present
```

```python
#Split-out validation dataset
array = df.values
#seperate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
df
```

```python
#TSNE visualization
data_tsne = TSNE(n_components=2).fit_transform(X)
cluster.tsneplot(score=data_tsne) #visualize data
```

```python
data_tsne
```

```python
newdf=pd.DataFrame(data_tsne).join(df['diagnosis'])
newdf
```

```python
#get a list categories
color_class = df['diagnosis'].to_numpy() #Convert Dataframe to array. legendpose=position on upper right
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos='upper right',legendanchor=(1.15,1))
#plt.show()
#plot will be stored in the default directory
```
