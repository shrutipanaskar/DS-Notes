```python
#import Hierarchical clustering libraries
import scipy.cluster.hierarchy as sch #to build dendrogram
from sklearn.cluster import AgglomerativeClustering

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
Univ = pd.read_csv("/content/Universities.csv")
```

```python
Univ
```

```python
#Customized normalization Function
#Here we can use standardized functions as well from sklearn
def norm_func(i): #function anme is norm_func, we can give any name here
  x = (i-i.min())/(i.max()-i.min()) #(Xi-min)/range
  return x
```

```python
#Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:]) # from 1st column because 0 index
```

```python
df_norm
```

```python
#create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
#create the clusters, suppose got input from customer that go for 4 clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage ='single')
```

```python
hc
```

```python
#save clusters for chart
y_hc = hc.fit_predict(df_norm) #apply fit_predict method on dataset df_norm
Clusters=pd.DataFrame(y_hc,columns=['clusters'])
```

```python
y_hc
```

```python
Clusters # Data point 0 belongs to 0th cluster
```

```python
# Now let us map this cluster membership to the data points
Univ['h_clusterid'] = Clusters
```
